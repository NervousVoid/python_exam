```python
def main(args):
    return str(
        (int(args[2], 10) << 20)
        + (int(args[1], 10) << 14)
        + (int(args[0], 10) << 0)
    )
 
 
print(main(('88', '1', '87'))) # '91242584'
print(main(('137', '57', '22'))) # '24002697'
print(main(('497', '17', '43'))) # '45367793'
print(main(('387', '53', '108'))) # '114114947'
```

# Задача №9

```python
def remove_none(x):
    res = []
    for row in x:
        if row[0] is None:
            continue
        temp = []
        for cell in row:
            if cell is None:
                continue
            temp.append(cell)
        res.append(temp)
    return res


def remove_duplicates(x):
    output = []
    control = set()
    for row in x:
        if row[0] in control:
            continue
        control.add(row[0])
        output.append(row)
    return output

def remove_duplicates_line(x):
    arr = []
    output = []
    for row in x:
        for y in row:
          if y not in output: 
            output.append(y)
        arr.append(output)
        output = []
    return arr



def formatt(res):
    for row in res:
        buf = row[0].split('&')
        row[0] = buf[1][:-6]
        row.append(buf[0].lower())
        row[1] = float(row[1])
        row[1] = "{:.2f}".format(row[1])
        
    return res


def transpose(A):
    C = []
    for i in range(len(A[0])):
        ci = []
        for j in range(len(A)):
            a = A[j][i]
            ci.append(a)
        C.append(ci)
    return (C)


def main(x):
    res = remove_none(x)
    res = remove_duplicates(res)
    res = remove_duplicates_line(res)
    res = formatt(res)
    res = sorted(res, key=lambda x: x[0])
    res = transpose(res)
    return res

print(main([['Да&Содянц, С.Т.', None, 0.9, None, 0.9], ['Да&Мицофман, Р.С.', None, 0.2, None, 0.2], ['Да&Содянц, С.Т.', None, 0.9, None, 0.9], ['Нет&Зунавук, Ф.К.', None, 0.9, None, 0.9]]))
```

```python
class MealyError(Exception):
    pass

class MealyMachine:
 def __init__(self):
    self.state = 'A'

 def view(self):
    if self.state == 'A':
        self.state = 'B'
        return 0
    elif self.state == 'B':
        self.state = 'C'
        return 1
    elif self.state == 'C':
        self.state = 'F'
        return 3
    elif self.state == 'E':
        self.state = 'B'
        return 7
    elif self.state == 'D':
        self.state = 'B'
        return 5
    else:
        raise MealyError("view")

 def reset(self):
    if self.state == 'F':
        return 8
    elif self.state == 'C':
        self.state = 'D'
        return 2
    elif self.state == 'D':
        self.state = 'E'
        return 4
    elif self.state == 'E':
        self.state = 'F'
        return 6
    else:
        raise MealyError("reset")

def main():
    return MealyMachine()

def raises(func, error):
    output = None
    try:
        output = func()
    except Exception as e:
        assert type(e) == error
    assert output is None


def test():
    o = main()
    raises(lambda: o.reset(), MealyError)
    assert o.view() == 0
    raises(lambda: o.reset(), MealyError)
    assert o.view() == 1
    assert o.view() == 3
    assert o.reset() == 8
    raises(lambda: o.view(), MealyError)
    assert o.reset() == 8
    o = main()
    raises(lambda: o.reset(), MealyError)
    assert o.view() == 0
    assert o.view() == 1
    assert o.reset() == 2
    assert o.view() == 5
    assert o.view() == 1
    assert o.reset() == 2
    assert o.reset() == 4
    assert o.reset() == 6
    assert o.reset() == 8
    raises(lambda: o.view(), MealyError)
    o = MealyMachine()
    raises(lambda: o.reset(), MealyError)
    assert o.view() == 0
    assert o.view() == 1
    assert o.reset() == 2
    assert o.reset() == 4
    assert o.view() == 7
    assert o.view() == 1
    assert o.view() == 3
    assert o.reset() == 8
    raises(lambda: o.view(), MealyError)
    assert o.reset() == 8
```

```python
from struct import unpack_from, calcsize
from typing import Any, Callable


class Types:
    char = 'c'
    int8 = 'b'
    uint8 = 'B'
    int16 = 'h'
    uint16 = 'H'
    int32 = 'i'
    uint32 = 'I'
    int64 = 'q'
    uint64 = 'Q'
    float = 'f'
    double = 'd'


class BinaryReader:
    def __init__(self, stream, offset, order=">"):
        self.stream = stream
        self.offset = offset
        self.order = order

    def jump(self, offset):
        reader = BinaryReader(self.stream, offset, self.order)
        return reader

    def read(self, pattern):
        size = calcsize(pattern)
        data = unpack_from(self.order + pattern, self.stream, self.offset)
        self.offset += size
        return data[0]


def read_d(reader):
    d1 = reader.read(Types.uint32)
    d2_size = reader.read(Types.uint32)
    d2_offset = reader.read(Types.uint32)
    d2_reader = reader.jump(d2_offset)
    d2 = [d2_reader.read(Types.uint64) for _ in range(d2_size)]
    return dict(D1=d1, D2=d2)


def read_c(reader):
    c1 = reader.read(Types.int8)
    c2 = reader.read(Types.int16)
    return dict(C1=c1, C2=c2)


def read_b(reader):
    b1 = reader.read(Types.int8)
    b2 = reader.read(Types.int8)
    b3_size = reader.read(Types.uint32)
    b3_offset = reader.read(Types.uint16)
    b3_reader = reader.jump(b3_offset)
    b3 = [b3_reader.read(Types.char) for _ in range(b3_size)]
    b3 = b"".join(b3).decode("utf-8")
    b4 = reader.read(Types.uint16)
    return dict(B1=b1, B2=b2, B3=b3, B4=b4)


def read_a(reader):
    a1 = read_b(reader)
    a2_size = reader.read(Types.uint32)
    a2_offset = reader.read(Types.uint32)
    a2_reader = reader.jump(a2_offset)
    a2 = [a2_reader.read(Types.char) for _ in range(a2_size)]
    a2 = b"".join(a2).decode("utf-8")
    a3 = [read_c(reader.jump(reader.read(Types.uint32))) for _ in range(2)]
    a4 = reader.read(Types.uint8)
    d_offset = reader.read(Types.uint32)
    d_reader = reader.jump(d_offset)
    a5 = read_d(d_reader)
    a6_size = reader.read(Types.uint32)
    a6_offset = reader.read(Types.uint16)
    a6_reader = reader.jump(a6_offset)
    a6 = [a6_reader.read(Types.int32) for _ in range(a6_size)]
    return dict(A1=a1, A2=a2, A3=a3, A4=a4, A5=a5, A6=a6)


def main(stream):
    return read_a(BinaryReader(stream, 4))


binary_data = (b'NET\xcdl\n\x00\x00\x00\x05\x00)\t\xa0\x00\x00\x00\x06\x00\x00\x00.\x00\x00'
               b'\x004\x00\x00\x007v\x00\x00\x00j\x00\x00\x00\x05\x00vsmlqvdxikdi\x8e{n1'
               b'm&R\x804\xef\xe0\xa1\xf2\xcb\xe6}\xcf\x05e\xe8\x99\x12\xdd\xc7-lN\xc5'
               b'\x7f\xdc\x84\xb5\xceL}\xb7\x10C\xbf\x175\x17t\xe91\xe0[\x1b\x06\x96!w'
               b'j\xe1\xddz\xb8\xee\x00\x00\x00\x06\x00\x00\x00:\xa8\xb8!\xeb\xe3AP\\R\x15'
               b'\xb9im6*\xf0\xcf\x97\xf6\xf5')

result = main(binary_data)
print(result)
```
